```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

```python
shotlogs = pd.read_csv('shot_logs.csv')
shotlogs[:10]
```

```python
#Total number of games played
total_games = len(shotlogs['GAME_ID'].unique())
total_games
```

```python
#Game reference table

gameref = shotlogs.groupby(['GAME_ID', 'MATCHUP', 'LOCATION', 'W'])[['SHOT_NUMBER']].count()
gameref[:5]
```

```python
splitfunc = lambda x: pd.Series([i for i in reversed(x.split('-'))])
matchup_split = shotlogs['MATCHUP'].apply(splitfunc)
matchup_split.rename(columns={0:'TEAMS',1:'DATE'},inplace=True)
# print matchup_split[:10]
```

```python
shotlogs = shotlogs.merge(matchup_split,left_index=True, right_index=True)
```

```python
shotlogs[:6]
```

```python
shotlogs.describe()
```

```python
# Analyzing performance during last 2 minutes of final quarter

import datetime as dt
shotlogs['GAME_CLOCK_TIME'] = pd.to_datetime(shotlogs['GAME_CLOCK'], format='%M:%S')
shotlogs['GAME_CLOCK_MINUTES']=shotlogs['GAME_CLOCK_TIME'].dt.minute
shotlogs['GAME_CLOCK_SECONDS']=shotlogs['GAME_CLOCK_TIME'].dt.second
mask1 = shotlogs['GAME_CLOCK_MINUTES']<3
mask2 = shotlogs['PERIOD']==4
shotlogs_last2minutes = shotlogs[mask1 & mask2]
shotlogs_last2minutes[:5]
```
